```python
import pandas as pd
pd.set_option('display.max_rows', 6)
```

```python
df = pd.read_csv('../shared-resources/time_series/time-series.csv', index_col=0, header=0, parse_dates=True)
```

```python
df.head()
```

```python
c = [c for c in df.columns if c.startswith('tsla')]
tsla = df[c]
```

```python
tsla.head()
```

```python
# testing data set: the up or down in the stock price for the next day
y = tsla.tsla_adj_close
y = y.diff()
y = y.iloc[1:]
y.index = range(len(y))
y.head()
```

```python
X1 = tsla.tsla_adj_close.diff()
X1.index = range(len(X1))
X1.name = 'diff_1'
X1.head()
```

```python
X2 = tsla.tsla_adj_close.diff(2)
X2.index = range(len(X2))
X2.name = 'diff_2'
```

```python
X3 = tsla.tsla_adj_close.diff(3)
X3.index = range(len(X3))
X3.name = 'diff_3'
X3.head()
```

```python
# need to put x1, x2, x3, and y together
Xy = pd.concat([X1, X2, X3, y],axis=1)
Xy = Xy.iloc[3:-1]
Xy.iloc[-3:]
```

```python
# 3.1 logistic regression
from sklearn.linear_model import LogisticRegression

# X1.iloc[1:].shape

logreg = LogisticRegression(class_weight='balanced')
logreg = logreg.fit(Xy[['diff_1', 'diff_2', 'diff_3']], Xy.tsla_adj_close)
# logreg.score(Xy
```
