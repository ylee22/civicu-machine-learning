## Numpy

```python
# import the numpy package
import numpy  # often also imported as np with `import numpy as np`
```

```python
# Create zero vectors
numpy.zeros(3)
```

```python
# Create zero matrix
numpy.zeros((3,2))
```

```python
# Transpose a matrix
numpy.transpose(numpy.zeros((3,2)))
```

```python
# dot product in numpy (multiply every vector element with the corresponding element in the other vector)
numpy.dot(numpy.ones(3), numpy.ones(3), )
```

```python
# other useful methods
numpy.eye(3)
```

```python
# another application of the dot product: every row of the first matrix is multiplied with the 
# vector and the sum will be the result for the 1st, 2nd, 3rd result vector element
numpy.dot(numpy.array(range(1, 4)), numpy.eye(3))
```

```python
# plot data with matplotlib
# line below is only needed in iPython notebooks
%matplotlib inline  
import matplotlib.pyplot as plt

print numpy.ones(10)
# random data
for symbol in ['rx', 'bo', '']:
    y = numpy.random.rand(10,1)
    x = numpy.ones(10) * numpy.array(range(1, 11))
    plt.plot(x, y, symbol)
    plt.ylabel('some random y numbers')
    plt.xlabel('some not so random x numbers')
    plt.show()
```

```python
# use the numpy where method
x = numpy.ones(10) * numpy.array(range(1, 11))
x[numpy.where( x > 3.0 )]
```
